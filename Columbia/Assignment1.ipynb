```python
import numpy as np
import astropy.units as units
```

```python
from astropy import constants as const
import matplotlib.pyplot as plt
```

```python
def my_planck(wavelength, temp):
    top = 2 * const.h * const.c**2 / wavelength**5
    bottom = np.exp((const.h * const.c) / (wavelength * const.k_B * temp)) - 1
    return(top / bottom).to(units.erg/units.s/units.cm**2/units.Angstrom)
    
```

```python
wave_arr = np.arange(2000, 14000, 10) * units.Angstrom
```

```python
Planck_5772 = my_planck(wave_arr, 5772*units.K)
Planck_7000 = my_planck(wave_arr, 7000*units.K)
Planck_4000 = my_planck(wave_arr, 4000*units.K)
```

```python
plt.figure(figsize=(8,6))

plt.plot(wave_arr, Planck_7000, label=r'T = 7000 K', c='blue')
plt.plot(wave_arr, Planck_5772, label=r'T = 5772 K', c='red')
plt.plot(wave_arr, Planck_4000, label=r'T = 4000 K', c='green')

plt.legend(fontsize='xx-large',edgecolor='black')

plt.xlabel('Wavelength (Angstroms)',fontsize=15)
plt.ylabel('Flux (erg s$^{-1}$ cm$^{-2}$ $\AA^{-1}$)',fontsize=15)

plt.show()
```
